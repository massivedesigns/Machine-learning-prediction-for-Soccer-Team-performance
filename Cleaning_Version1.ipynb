```python
def duplicate_soln(df):
    
    df = df.copy()
    
    # Checking the shape of the DataFrame after attempting to drop duplicates
    print(f'Shape after dropping duplicates: {df.shape}')
    
    # Convert float columns to a rounded format to avoid floating-point discrepancies
    float_cols = df.select_dtypes(include=[np.float64]).columns
    df[float_cols] = df[float_cols].round(decimals=3)

    # Drop duplicates based on a logical subset of columns including 'Opponent'
    key_columns = ['Player', 'Date', 'Comp', 'Round', 'Opponent']
    df_deduped = df.drop_duplicates(subset=key_columns)

    # Sort the DataFrame by 'Date'
    df_deduped = df_deduped.sort_values(by='Date', ascending=True)

    # Print the shape of the DataFrame after processing
    print(f'Shape after dropping duplicates: {df_deduped.shape}')

    return df_deduped
```

```python


def add_player_roles(df, position_column):
    # Define position categories
    defensive_positions = {'CB', 'LB', 'RB', 'WB'}
    forward_positions = {'FW', 'RW', 'LW'}
    midfield_positions = {'CM', 'DM', 'AM', 'RM', 'LM'}
    goalkeeper_position = 'GK'

    # Helper function to determine the most common role
    def find_most_common_role(positions):
        role_count = {'defense': 0, 'midfield': 0, 'forward': 0, 'goalkeeper': 0}
        for pos in positions:
            if pos == goalkeeper_position:
                return 'goalkeeper'  # Immediate return if position is GK
            if pos in defensive_positions:
                role_count['defense'] += 1
            elif pos in midfield_positions:
                role_count['midfield'] += 1
            elif pos in forward_positions:
                role_count['forward'] += 1

        return max(role_count, key=role_count.get)

    # Determine the most common role for each player
    def player_common_role(group):
        positions = group[position_column].str.split(',').explode().str.strip()
        return find_most_common_role(positions)

    # Calculate the most common role for each player and create a new column
    player_roles = df.groupby('Player').apply(player_common_role).reset_index(name='Role')
    df = df.merge(player_roles, on='Player', how='left')

    return df

```
